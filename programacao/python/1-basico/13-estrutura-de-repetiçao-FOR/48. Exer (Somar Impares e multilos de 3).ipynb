Exercício Python 048: Faça um programa que calcule a soma dos números impares multiplos de 3 dentro de um intervalo escolhido pelo usuario.

```python
print('''\n
Digite o inicio e o fim de um intervalo para a maquina calcular
somente a soma dos números ímpares multiplos de 3.\n''')

i = int(input("Inicio: " ))
f = int(input("Fim: " ))
print()

som = cont = 0

for c in range(i,f+1):
    if c%2==1:
        if c%3==0:
            print(c)
            som = som+c
            cont = cont + 1

print(f"\nA soma de todos os {cont} valores solicitados é {som}")
```
