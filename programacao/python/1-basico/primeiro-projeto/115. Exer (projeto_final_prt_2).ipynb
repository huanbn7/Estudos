```python
from lib import funcoes as fn
import time as tm


opcoes = ['Ver pesssoas cadastradas', 'Cadastrar novas pessoas', 'Sair do sistema']


while True:
    fn.menu(opcoes)
        
    resposta = fn.valida('Sua opcao: ')
    
    if resposta == '':
        tm.sleep(0.5)
        continue
        
    elif resposta == 1:
        fn.cabecalho('OPCAO 1')
        fn.ver_arquivo()
        
    elif resposta == 2:
        fn.cabecalho('OPCAO 2')
        while True:
            saida = fn.cadastrar('Nome: ')
            print()

            if saida == 'n':
                break
        
    elif resposta == 3:
        fn.cabecalho('Saindo do sistema... Ate logo!')
        break
    
    else:
        fn.colorir('Error: Digite uma opcao valida!',0)
    
    tm.sleep(0.5)
```
