```python
from time import *

tupla=('Corinthians','Palmeiras','Grêmio','Santos','Cruzeiro','Flamengo','Vasco da Gama','Botafogo','Chapecoense','Atlético-MG','Bahia','São Paulo','Athletico','Fluminense','Vitória','Coritiba','Sport Recife','Avaí','Ponte Preta','Atlético-GO')


while True:
    print('''▶ O QUE VOCÊ QUER SABER SOBRE O BRASILEIRÃO? 
    
[1] Tabela completa
[2] Os times classificados para a LIBERTA
[3] Os times Rebaixados
[4] Quem foi o campeão Brasileiro
[5] Tabela em ordem alfabetica
    ''')
    
    num = int(input('▶ DIGITE UM NUMERO: '))
    sleep(1.3)
    while num > 5 or num < 1:
        num = int(input('▶ ERROR. DIGITE UM NUMERO DE 1 Á 5:'))
        sleep(1.3)
    print()
    
    
    if num == 1:
        print('\033[43;1;34mTABELA BRASILEIRÃO 2017\033[m\n')
        
        for pos,time in enumerate(tupla):
            print(f'\033[1m{pos+1}°\033[m | \033[1;34m{time}\033[m')
        print()
        
    
    if num == 2:
        print('\033[43;1;34mTIMES CLASSIFICADOS PARA A LIBERTA\033[m\n')
        for pos,time in enumerate(tupla[0:5]):
            print(f'\033[1m{pos+1}°\033[m | \033[1;34m{time}\033[m')
        print()
        
    
    if num == 3:
        print('\033[43;1;34mTIMES EBAIXADOS PARA A SÉRIE B\033[m\n')
        for pos,time in enumerate(tupla[16:21]):
            print(f'\033[1m{pos+17}°\033[m | \033[1;31m{time}\033[m')
        print()
    
    if num == 4:
        print(f'\033[1;33mO campeão do Brasileirão 2017 foi o\033[m \033[1;34m{tupla[0]}\033[m.')
    
    if num == 5:
        print('''\033[43;1;34mTABELA BRASILEIRÃO 2017 EM ORDEM ALFABÉTICA\033[m\n''')
        
        for pos,time in enumerate(sorted(tupla)):
            print(f'\033[1m{pos+1}°\033[m | \033[1;34m{time}\033[m')
        print()
     
    
    #saida
    print()
    print()
    saida = input('▶ QUER CONTINUAR, SIM OU NÃO? ').casefold().strip()[0]
    sleep(0.8)
    while saida != 's' and saida != 'n':
        saida = input('▶ ERROR. DIGITE S PARA CONTINUAR E N PARA PARAR: ').casefold().strip()[0]
        sleep(0.8)
    
    if saida == 'n':
        break
    
```
