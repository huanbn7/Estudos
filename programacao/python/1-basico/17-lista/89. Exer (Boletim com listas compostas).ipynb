Exercício Python 089: Crie um programa que leia nome e duas notas de vários alunos e guarde tudo em uma lista composta. No final, mostre um boletim contendo a média de cada um e permita que o usuário possa mostrar as notas de cada aluno individualmente.

```python
lista = []

#coletar e adicionar os dados
while True:
    dados = [input('\nNome: ').strip().capitalize(),int(input('Nota 1: ')),int(input('Nota 2: '))]
    lista.append(dados[:])
    dados.clear()
    
    if input('Quer continuar [s/n]? ') in 'Nn':
        break

#mostrar uma lista com os dados
print()
print('='*28)
print(f'  {"N°":<6}{"Nome":<12}{"Média"}')
print('='*28)


for sl in lista:
    media = (sl[1]+sl[2])/2
    print(f'  {lista.index(sl)+1:<6}{sl[0]:<12}{media}')

print('='*28) 


#mostrar nosta individualmente
while True:
    aluno = int(input('\nMostrar notas de qual aluno [99 interrompe]? '))
    if aluno == 99:
        print('\nPrograma finalizado.')
        break
    
    for sl in lista:
        if -1 < aluno <= len(lista):
            print(f'As notas de {lista[aluno-1][0]} são: {lista[aluno-1][1]} e {lista[aluno-1][2]}.')
            break
        else:
            print('Por favor veja se o numero do aluno está correto.')
            break
    

              
```
