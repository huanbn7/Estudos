## Heap

#### Introduçao

    O heap é uma estrutura de dados que segue as propriedades de uma arvore binaria completa, ou seja, um heap tem todos os niveis preenchidos e esses niveis sao preenchidos de cima para baixo da esquerda para direita. E satifaz a propridade de min-heap ou max-heap, ou seja: 
    
    - min-heap: cada no pai deve ser menor ou igual ao seu(s) filho(s). 
    - max heap: cada no pai deve ser maior ou igual a seu(s) filho(s).
    
    As formulas para saber as relacoes dos elementos e fazer muitas outras coisas sao:
    
    - Se um pai esta no indice i
    - Seu filho esquerdo esta no indice 2*i+1
    - Seu filho direito esta no indice 2*i+2
    
    Para saber o pai de um elemento (indice do elemento - 1) // 2 

    Obs: Nao existe uma modulo no python para criar e manipular um max-heap diretamente, mas é possivel fazer esses processos deixando todos os valores de um heap negativo e na hora de remover esse valor transformar ele em positivo, outra maneira é criar uma class que tenha propriedades para criar e manipular um max-heap. 

#### Operacoes basicas

    heapify
    
    O processo heapify é ultilizado para transformar iterables normais em um min-heap ou max-heap, utiliza-se as formulas citadas na introduçao para descobrir as relaçoes de pai e filho e comparando o pai e seus filhos para organizar conforme sua propriedade de min ou max heap.

    heappush
    
    Processo de inserir um elemento no heap.

    heappop
    
    Processo para capturar o menor ou o maior elemento do heap, no caso o primeiro elemento do vetor.

    heapreplace
    
    Esse processo remove o dado com menor valor e susbstituí por outro.

#### Complexidade de tempo 

    A complexidade de tempo de execuçao do heap de contruçao é O(n log(n)). Onde cada chamada para o heapify custa O (log(n)) e o custo de construir o heap é O(n). Postanto a complexidade de tempo total sera de O(n log(n)).

#### Aplicaçoes do heap

    - O heap é usado ao implementar a fila de prioridade
    - Heap é usado na classificaçao de heap
    - A estrutura de dados heap é usado ao trabalhar com o algoritmo de Dijkstra
    - Podemos usar um min-heap ou um max-heap no sistema operacional para o algoritmo de agendamento de tarefas
    - é utilizado no algoritmo de seleçao
    - A estrutura de dados heap é usada em algoritmos de grafico como o algoritmo de prim
    - é usado em estatistica de pedidos
    - A estruta de dados heap é usada na mesclagem k-way

## Modulo heapq

```python
import heapq 
```

    Esse modulo fornece uma implementaçao do algoritimo de fila heap, tambem conhecido como algoritimo de prioridade.

#### Funcoes para criacao e manipulacao de heap

#### heapify(x)

    Transforma uma lista em um heap, no local, em tempo linear.

```python
a = [1, 4, 8, 0, -1]
heapq.heapify(a)

print(a)
```

heappush(heap, item)

    Adiciona um elemento ao heap. Nao se deve aplicar essa funcao a uma estrutura de dados que nao seja um heap, pois a ordem da estrutura de dados nao vai estar como o desejado.



```python
a = [1, 4, 8, 0, -1]
heapq.heapify(a)

heapq.heappush(a, -2)
heapq.heappush(a, 2)

print(a)
```

heappop(heap)

    é usado para remover o menor item que sempre estara no indice 0. Além disso garante que o ultimo elemento seje colocado na posicao do elemento removido, e apos esse processo o item que ocupa a primeira posicao da array deve ser compado com seus filhos até que encontre sua pocisao conforme as propriedades de um heap.

```python
a = [1, 4, 8, 0, -1]
heapq.heapify(a)
print(a)

heapq.heappop(a)
print(a)
```

heappushpop(heap, item)

    Essa funcao primeiro adiciona o item fornecido em um heap tambem fornecido. Entao remove o menor e o retorna, ou seja, é a junçao da funçao heappush() e heappop(). 

```python
a = [1, 4, 8, 0, -1]
heapq.heapify(a)
print(a,'\n')

b = heapq.heappushpop(a, 100) # insere o 100, exclui e retorna o menor valor.
print(b, '\n')
print(a)
```

heapreplace(heap, item)

    Exclui o menor elemento do heap e em seguida insere um novo item fornecido. Essa funcao é mais eficiente do que chamar heappop() e heappush().

```python
a = [1, 4, 8, 0, -1]
heapq.heapify(a)
print(a,'\n')

b = heapq.heapreplace(a, 100) # insere o 100, exclui e retorna o menor valor.
print(b, '\n')
print(a)
```

nlargest(n, iterable, key=None)

    Encontra os n's numeros maiores de uma determinada iterable. Essa funçao tambem aceita uma chave que é a funçao de um argumento, que é usado para extrair uma chave de comparaçao de cada elemento de uma iterable.

```python
a = [1, 4, 8, 0, -1]
heapq.heapify(a)
print(a,'\n')

b = heapq.nlargest(n=2, iterable=a, key=None)
print(b, '\n')
print(a)
```

nsmallest (n, iterable, key=None)

    Mesma coisa do nlargest() so que agora retorna uma lista com os menores valores.

    Obs: Quando n == 1, é mais eficiente usar as funçoes internas min() e max(). 

```python
a = [1, 4, 8, 0, -1]
heapq.heapify(a)
print(a,'\n')

b = heapq.nsmallest(n=2, iterable=a, key=None)
print(b, '\n')
print(a)
```

## Explicar o que e uma fila de prioridades

#### Introduçao 
 
    As filas de prioridades sao estruturas de dados abstratos onde cada dado/valor na fila tem uma certa prioridade.
    Essa estrutura de dados é usada para lidar com problemas de agendamento em que algumas tarefas sao priorizadas em releçao a outras. 
    Por exemplo, em um hospital um paciente com traumatismo craniano tem mais prioridade do que um paciente com dor de cabeça.
    Um a maneira de implementar essa estrutura de dados abstrata é usando heap.


#### Propriedades

    1. Um elemento com alta prioridade é desinfeleirado antes de um elemento com uma prioridade mais baixa.
    2. Se dois elementos tiverem os mesmos valores, eles serao atendido de acordo com sua ordem na fila.
    

#### Aplicaçoes

    - Algoritmo de job scheduling
    - CPU e Disk Scheduling
    - Gerenciamento de recursos que sao compartilhados entre diferente processos
    - Etc

#### Exemplos


```python
fila_hospital = list()
heapq.heapify(fila_hospital)

heapq.heappush(fila_hospital, (1,'Traumatismo craniano'))
heapq.heappush(fila_hospital, (2,'Convusao/Febre 40 graus'))
heapq.heappush(fila_hospital, (3,'Dor de cabeça'))
heapq.heappush(fila_hospital, (1,'Parada cardiaca'))
heapq.heappush(fila_hospital, (3,'Dor de barriga'))
heapq.heappush(fila_hospital, (1,'Tiro na cabeça'))


print(fila_hospital)
```
